```python
import pandas as pd

# Lista de file_ids de los archivos públicos en Drive
file_ids = [
    "1yD28PmPmByNvtJAlZ7JchKuOss_oUYv_",    #mayo
    "1WnnLypdzIi9whk54FwpXG2ersSyGlnhT",    #junio
    "1aoQngs8yK65wTd8Kqqkw_K5lCKAUKQEz",    #julio
    "1pkNHi90yDIPVwjN_5QhQAgqwilqxeyow",    #agosto
    "1Duiq8PFAaSf8adVdWZPJvUcBxKlLyo_N",    #septiembre
    "1ol1iZJiVlM8Ova8EQjByvLNk01Lzu6Ph",    #octubre
    "1xnvdKLA8r3PSkB8UvEQq2VHGVHG8JnTC"     #noviembre
]
dfs = []

for file_id in file_ids:
    try:
        url = f"https://drive.google.com/uc?id={file_id}"
        df = pd.read_excel(url, sheet_name="AGU02 PERFIL", engine="openpyxl", header=1)
        dfs.append(df)
    except Exception as e:
        print(f"Error al leer archivo con ID {file_id}: {e}")

# Unir todos los DataFrames
df_total = pd.concat(dfs, ignore_index=True)

# Ver los primeros registros
print(df_total.head())

```

```python
# Eliminamos la columna 2 y 4 debido a que no contiene datos relevantes para el estudio
df_total = df_total.drop(columns=[df_total.columns[1], df_total.columns[3]])
```

```python
# nos quedamos con las siguientes columnas ya que contiene los datos utiles para el estudio
df_total = df_total.iloc[:, :17]
```

```python
# cambiamos a tipo de dato datetime la columna porque esta contiene las fechas
df_total['Unnamed: 2'] = pd.to_datetime(df_total[df_total.columns[1]], errors='coerce')
```

```python
# renombramos a la columna que contiene las fechas
df_total = df_total.rename(columns={df_total.columns[1]: 'Fecha'})
```

```python
# eliminamos los valores nulos que estan en la columna fecha
df_total = df_total.dropna(subset=['Fecha'])
```

```python
# Convertir todas las demás columnas numéricas
columnas_numericas = df_total.columns.drop(['Fecha'])

for col in columnas_numericas:
    df_total[col] = pd.to_numeric(df_total[col], errors='coerce')
```

```python
# Creamos una nueva columna para buscar fechas que pueden estar repetidas y poder hacer la eliminacion de duplicados
df_total = df_total.sort_values('Fecha')  # Ordena
df_total['Delta'] = df_total['Fecha'].diff() # Creamos una columna que contiene la diferencia de cada fila y la anterior
print(df_total['Delta'].value_counts())
```

```python
# Eliminar las filas con diferencia de 0 minutos entre fechas consecutivas
df_total = df_total[df_total['Delta'] != pd.Timedelta(0)]

# Eliminar la columna auxiliar
df_total = df_total.drop(columns=['Delta'])

# Verificar resultado
print(df_total.shape)
```

```python
# Renombramos las columnas
df_total.columns = [
    'nro',
    'Fecha',
    'Potencia_activa_generada [kW]',
    'Potencia_activa_consumida [kW]',
    'Potencia_reactiva_generada [kVAr]',
    'Potencia_reactiva_consumida [kVAr]',
    'Potencia_aparente_generada [kVA]',
    'Potencia_aparente_consumida [kVA]',
    'Voltaje_F-N_A [V]',
    'Voltaje_F-N_B [V]',
    'Voltaje_F-N_C [V]',
    'Amperaje_A [A]',
    'Amperaje_B [A]',
    'Amperaje_C [A]',
    'Factor_potencia_generado',
    'Factor_potencia_consumido',
    'Frecuencia'
]
```

```python
df_total = df_total.sort_values('Fecha').reset_index(drop=True)
```

```python
# creamos una nueva columna que funcionara como indice
df_total['Nro_evento'] = range(1, len(df_total) + 1)
```

```python
# eliminamos la columna Nro
if 'nro' in df_total.columns:
    df_total = df_total.drop(columns=['nro'])
```

```python
# reordenamos las columnas
columnas = ['Nro_evento'] + [col for col in df_total.columns if col != 'Nro_evento']
df_total = df_total[columnas]
```

```python
print(df_total.info())
```

```python
# Buscamos fechas en las cuales los registros sean diferentes a los 15 min
horas_incorrectas = df_total[
    (df_total['Fecha'].dt.minute % 15 != 0) |
    (df_total['Fecha'].dt.second != 0)
]

# Mostrar los registros conflictivos
print(horas_incorrectas[['Fecha']])
print(f"\nCantidad de registros con horario incorrecto: {len(horas_incorrectas)}")
```

```python
# Redondear toda la columna 'Fecha' al múltiplo más cercano de 15 minutos
df_total['Fecha'] = df_total['Fecha'].dt.round('15min')
```
